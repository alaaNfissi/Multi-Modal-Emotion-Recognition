```
import os
import pickle


# Step 1: Load the data from meta.pkl

file_path='/content/meta.pkl' #replace the path

with open(file_path, 'rb') as f:
    data = pickle.load(f)
```

```

# Replace label 'exc' with 'hap'
for key, value in data.items():
    if value['label'] == 'exc':
        value['label'] = 'hap'



# Keep only the items with labels in ['hap', 'ang', 'sad', 'neu'] from keys_list
valid_labels = {'hap', 'ang', 'sad', 'neu'}

with open(file_path, 'wb') as f:
    pickle.dump(data, f)

print(f"Data has been updated and saved back to {file_path}")
```

```
#run this code for train_split, val_split, test_split.

# Load items from train_split.txt
items_path = "train_split.txt"  #replace the file path

with open(items_path, "r") as file:
    items_list = [line.strip() for line in file.readlines()]


filtered_items = [key for key in items_list if data.get(key, {}).get('label') in valid_labels]

# Save the filtered items back to file.txt
with open(items_path, "w") as file:
    file.writelines([key + "\n" for key in filtered_items])

print(f"Filtered items have been saved back to {items_path}")
```
